```python
file_path = "/Users/hanifrawi/Desktop/shopee/shopee_project.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
import pandas as pd

# Convert 'w_date' column to datetime format
df['w_date'] = pd.to_datetime(df['w_date'], errors='coerce')

# Filter for May 2023
df_may = df[(df['w_date'] >= '2023-05-01') & (df['w_date'] <= '2023-05-31')]

# Display the first few rows
df_may.head()
```

```python
df_may.groupby('w_date').size().plot(kind='line')
```

```python
import re
import pandas as pd

# List of Malaysian states
states = ["Selangor", "Johor", "Melaka", "Kedah", "Penang", "Pahang", "Sabah", "Sarawak"]

# Extract location from 'specification' column
df_may['location'] = df_may['specification'].astype(str).apply(lambda x: next((state for state in states if state in x), "Oversea"))

# Show results
df_may[['specification', 'location']].head()
```

```python
# Extract main category safely
df_may['Main_Category'] = df_may['item_category_detail'].astype(str).apply(lambda x: x.split('|')[1].strip() if '|' in x else 'Unknown')

# Count occurrences of each category
print(df_may['Main_Category'].value_counts())
```

```python
# Ensure 'price_actual' is numeric
df_may['price_actual'] = pd.to_numeric(df_may['price_actual'], errors='coerce')

# Group by 'Main_Category' and calculate min, max, median prices
df_may.groupby('Main_Category')['price_actual'].agg(['min', 'max', 'median'])
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Ensure 'total_sold' and 'price_actual' are numeric
df_may['total_sold'] = pd.to_numeric(df_may['total_sold'], errors='coerce')
df_may['price_actual'] = pd.to_numeric(df_may['price_actual'], errors='coerce')

# Calculate revenue
df_may['revenue'] = df_may['total_sold'] * df_may['price_actual']

# Group by 'Main_Category' and plot revenue
df_may.groupby('Main_Category')['revenue'].sum().sort_values(ascending=False).plot(kind='bar', color='skyblue')

# Show the plot
plt.xlabel("Main Category")
plt.ylabel("Total Revenue")
plt.title("Total Revenue by Main Category")
plt.xticks(rotation=45)
plt.show()
```
